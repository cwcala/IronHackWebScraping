```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
```

```python
#save target url
url = 'https://www.imdb.com/search/title/?groups=top_250&sort=user_rating'
```

```python
response = requests.get(url)
```

```python
html = response.content
```

```python
#parser lxml 
soup = BeautifulSoup(html,'lxml')
```

To inspect parts of web-page to find html tags that correspond to the data we want to extract on the target web page, it can be helpful to install the chrome extension, CSS Selector Tester https://chrome.google.com/webstore/search/css%20selector%20tester?hl=fr which is like the chrome inspect mode on steroids, since it highlights relevant parts of the web page corresponding to various CSS selectors, so you can then use BeautifulSoup select with the right CSS selector to get the right content.

Now to use the various CSS selectors to isolate the features we want, such as number of votes, name of film, rating, year, metascore,... 

```python
#isolate the votes using the p tag, the votes class and the 2nd child as there are several of them
votes = [i.text for i in soup.select("p.sort-num_votes-visible :nth-child(2)")]
```

```python
#isolate the year using the span tag and the lister-item-year class
year = [i.text for i in soup.select('span.lister-item-year')]
```

```python
#isolate the names using the h3 tag and the lister-item-header class and a tag
#for i in soup.select('h3.lister-item-header a'):
 # print(i.text)
Name = [i.text for i in soup.select('h3.lister-item-header a')]
```

```python
#isolate the metascore using the div tag, inline-block class and span tag with metascore
metascore = [i.text for i in soup.select('div.inline-block span.metascore')]
```

```python
#put the 50 blocks into a container
container = soup.select('div.lister-item-content')
```

Now to do the same thing for the first several pages of results, not just the first one by iterating over page number k in pages as a string interpolation in the url.  The time.sleep puts a second delay between iterations to avoid oversoliciting the website.  Each page of results for each feature is saved into the new feature1 lists.

```python
#iterate for the first 5 pages of 50 results
pages = [1,51,101,151,201]
name1 = []
year1 = []
rating1 = []
votes1 = []
met1 = []
for k in pages:
    url = f'https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start={k}&ref_=adv_nxt'
    response = requests.get(url)
    time.sleep(1)
    html = response.content
    soup = BeautifulSoup(html, 'lxml')
    name = [i.text for i in soup.select('h3.lister-item-header a')]
    name1.append(name)
    year = [i.text for i in soup.select('span.lister-item-year')]
    year1.append(year)
    rating = [i.text for i in soup.select('div.inline-block strong')]
    rating1.append(rating)
    votes = [i.text for i in soup.select("p.sort-num_votes-visible :nth-child(2)")]
    votes1.append(votes)
    container = soup.select('div.lister-item-content')
    #metascore = [i.text for i in soup.select('div.inline-block span.metascore')]
    for i in container:
        metascores1 = [i.text for i in i.select('div.inline-block span.metascore')]
        if len(metascores1) == 0:
            met1.append(0)
        else:
            met1.append(metascores1)
```

```python
#Make final feature lists for each feature 
final_name = [j for i in name1 for j in i]
final_year = [j for i in year1 for j in i]
final_rating = [j for i in rating1 for j in i]
final_votes = [j for i in votes1 for j in i]
```

```python
#Build the final dataframe by adding each feature list as a column
final_df = pd.DataFrame()
final_df['Name'] = final_name
final_df['Year'] = final_year
final_df['Rating'] = final_rating
final_df['Votes'] = final_votes
final_df['Metascore'] = met1

final_df
```
